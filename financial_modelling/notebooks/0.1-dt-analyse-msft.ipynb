```python
import json
import pandas as pd
import pathlib
import plotly.express as px
```

```python
json_paths = ["../data/MSFT_20200210213749.json", "../data/AAPL_20200210222048.json"]
dfs = []
for jp in json_paths:
    with open(jp, "r") as f:
        raw_data = json.load(f)
        temp_df = (pd
                   .DataFrame(raw_data["chart"])
                   .assign(stock=raw_data["quote"]["symbol"])
                   )
    dfs.append(temp_df)
df = pd.concat(dfs).reset_index(drop=True)
```

```python
df
```

```python
fig = px.line(df, x="date", y="close", color="stock")
fig.show()
```
